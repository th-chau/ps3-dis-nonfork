### Quick check 2

Evaluate the following claim: 'weighing eliminates sampling biases'. Explain

**Solution:** False. Weighing can only 'weigh out' the variables that we account for. There are many other possible omitted variables, such as income, partisanship and so on.

### Quick check 3

What are some sources of sampling biases?

**Solution:** 
- Sample selection bias: The kind of people selected to be in the sample are systematically different than the population.
- Non-response bias: The kind of people who agree to respond to a survey are systematically different than population sampled

### Quick check 5

Evaluate the following sentence: 'since the p-value of the results above is smaller than 0.05 and the estimate is negative, being college educated causes one to be less supportive of Trump.'

**Solution:** False. The p-value being smaller than 0.05 means that the difference we see wouldn't be due to noise from sampling (i.e. there is a difference in Trump support among college educated voters in the population given random sampling), but it does not mean that being college educated has a *causal* effect
