Now, we will look a set of data on life expectancy and GDP per capita in 2015.

```r
life_gdp <- read.csv("./data/w10.csv")
head(life_gdp)
```

### Quick check 1

Plot the life expectancy against the GDP per capita below.

```r
library(ggplot2)
qplot(gdp_per_capita, expectancy, data = life_gdp)
```

### Quick check 2

Do the same plot but add the line of best fit

```r
qplot(gdp_per_capita, expectancy, data = life_gdp) + geom_smooth(method = 'lm')
```

### Quick check 3

Obtain the formula for the straight line above, by running a regression of `gdp_per_capita` against `expectancy`

```r
summary(lm(expectancy ~ gdp_per_capita, data = life_gdp))
```

### Quick check 4

Write the formula of the line below.

**Solution:** $expectancy = 67.4 + 0.0002458 \times gdp_per_capita$
